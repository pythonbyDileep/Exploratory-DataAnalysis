```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="gender", hue="gender", palette="pastel", legend=False)
plt.title("Gender Distribution")
plt.show()
```

```python
# Final Result Distribution
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="final_result", hue="final_result", palette="Set2", legend=False)
plt.title("Pass/Fail Distribution")
plt.show()
```

```python
# Score Distributions
score_cols = ["math_score", "reading_score", "writing_score"]
df[score_cols].hist(bins=20, figsize=(12, 6), color="#00bcd4")
plt.suptitle("Score Distributions")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.select_dtypes(include=["float64", "int64"]).corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Scores by Final Result
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="final_result", y="math_score", hue="final_result", palette="Set3", legend=False)
plt.title("Math Score by Final Result")
plt.show()
```

```python
# Study Hours vs Scores
sns.scatterplot(data=df, x="study_hours", y="math_score", hue="final_result")
plt.title("Study Hours vs Math Score")
plt.show()
```

```python
# Internet Access vs Final Result
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="internet_access", hue="final_result", palette="Set1")
plt.title("Internet Access and Final Result")
plt.show()
```

```python
#parrent education analysis
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x="parent_education", y="reading_score",
            hue="parent_education", palette="cool", legend=False)
plt.xticks(rotation=45)
plt.title("Reading Score vs Parent Education Level")
plt.show()
```
